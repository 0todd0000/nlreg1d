# Datasets

Four CSV datasets are available in `nlreg1d`:

- `SimulatedA`
- `SimulatedB`
- `Besier2009VastusForce`  (Besier et al., 2009)
- `Pataky2014MediolateralCOP`  (Pataky et al., 2014)

The CSV files associated with these datasets are available in the `Data` directory of this repository. Note that **the first column of each CSV file contains group integer labels**; the remaining columns contain the 1D dependent variable.

They can most conveniently be loaded and visualized as follows:

```python
import numpy as np
from matplotlib import pyplot as plt
import nlreg1d as nl
```

```python
# load:
dataset0  = nl.data.SimulatedA()
dataset1  = nl.data.SimulatedB()
dataset2  = nl.data.Besier2009VastusForce()
dataset3  = nl.data.Pataky2014MediolateralCOP()

# plot:
fig,axs   = plt.subplots(2, 2, figsize=(10,8))
for ax,dset in zip( axs.ravel(), [dataset0,dataset1,dataset2,dataset3]):
    dset.plot(ax=ax)
plt.show()
```

Printing the `dataset` object shows the file path, dependent variable (DV) array size, and unique group labels:

```python
print( dataset0 )
```

Key dataset attributes / properties include:

- `J`:  number of observations: (int)
- `Q`:  [int] number of grid points: (int)
- `dv`:  dependent variable: (J,Q) array
- `group`: group labels: (J,) array of integers

These can be accessed as follows:

```python
print( dataset0.J )
print( dataset0.Q )
print( dataset0.dv.shape )
print( dataset0.group )
```

Note that the transpose operator should be applied to the DV when plotting:

```python
ax = plt.axes()
y  = dataset0.dv
ax.plot( y.T )
plt.show()
```
